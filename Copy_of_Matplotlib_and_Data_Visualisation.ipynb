```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
plt.title('Line Plot with Custom Axes Labels and Limits')
plt.show()
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')

axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Scatter Plot')

axs[1, 0].bar(x[:10], y[:10])
axs[1, 0].set_title('Bar Plot')

axs[1, 1].hist(y, bins=10)
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.title('Line Plot with Legend')
plt.show()
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot Saved as PNG')
plt.savefig('line_plot.png')

plt.figure(figsize=(10, 6))

# Line Plot
plt.subplot(2, 1, 1)
plt.plot(x, y, label='Line Plot')
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# Area Plot
plt.subplot(2, 1, 2)
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title('Area Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.tight_layout()
plt.show()


import seaborn as sns
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

plt.figure(figsize=(12, 6))

# Scatter Plot
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='viridis')
plt.title('Scatter Plot')

# Bubble Plot
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', size='petal length (cm)', hue='species', palette='viridis', sizes=(20, 200))
plt.title('Bubble Plot')

plt.tight_layout()
plt.show()


# Simple Bar Plot
data = pd.DataFrame({'Categories': ['A', 'B', 'C'], 'Values': [10, 20, 30]})
plt.bar(data['Categories'], data['Values'])
plt.title('Simple Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Grouped Bar Plot
data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B'],
    'Subcategory': ['X', 'Y', 'X', 'Y'],
    'Value': [10, 20, 15, 25]
})
sns.barplot(x='Category', y='Value', hue='Subcategory', data=data)
plt.title('Grouped Bar Plot')
plt.show()

# Stacked Bar Plot
data = pd.DataFrame({
    'Category': ['A', 'B'],
    'X': [10, 15],
    'Y': [20, 25]
})
data.set_index('Category').plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot')
plt.ylabel('Values')
plt.show()


#Histogram
plt.hist(df['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()


#Pie plot
sizes = [10, 20, 30]
labels = ['A', 'B', 'C']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()

#Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.show()

#Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=df)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

#Marginal plot
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=df, kind='scatter')
plt.show()

#Contour plot
from scipy.stats import kde

x = df['sepal length (cm)']
y = df['sepal width (cm)']
k = kde.gaussian_kde([x, y])

xgrid, ygrid = np.mgrid[x.min():x.max():100j, y.min():y.max():100j]
z = k(np.vstack([xgrid.ravel(), ygrid.ravel()]))
z = z.reshape(xgrid.shape)

plt.contourf(xgrid, ygrid, z, cmap='Blues')
plt.colorbar()
plt.title('Contour Plot')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

#Heat map
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Features')
plt.show()


#3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'], c=df['species'], cmap='viridis')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
plt.title('3D Scatter Plot')
plt.show()

#Multivariate Distribution and Clustering
from sklearn.cluster import KMeans
import seaborn as sns

# KMeans clustering
kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])

# Plot clusters
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df, palette='viridis')
plt.title('KMeans Clustering')
plt.show()








```
